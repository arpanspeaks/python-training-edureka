# Q1. You are given a dataset, which is present in the LMS, containing the number of hurricanes occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, taking the Year as the X-axis and the number of hurricanes occurring as the Y-axis.

# Q2. The given dataset records the data of city temperatures over the years 2014 and 2015. Plot the histogram of the temperatures over this period for the cities of San Francisco and Moscow.

```python
df = pd.read_csv("CityTemps.csv")
print(df.head())

mosco = df["Moscow"]
sf = df["San Francisco"]

plt.hist(mosco, len(mosco))
plt.hist(sf, len(sf))

plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Temperatures of Mosco and San Francisco")

plt.show()
```

# Q3. Plot a pie-chart of the number of models released by every manufacturer, recorded in the provided data. Also, mention the name of the manufacturer with the largest releases.

```python
df = pd.read_csv("Car_sales.csv")
df = df[['Manufacturer', 'Sales in thousands']]
df1 = df.groupby('Manufacturer').agg('sum');
df1.reset_index(inplace=True)
df1.sort_values(by='Sales in thousands', ascending=False, inplace=True)
plt.figure(figsize=(60,12))
ax1 = plt.subplot(121, aspect='equal')
df1.plot(kind='pie', y='Sales in thousands', ax=ax1, autopct='%1.1f%%', startangle=90, shadow=False, labels=df1['Manufacturer'], legend = False, fontsize=10) 

plt.show()
```

### Q5.9: Create a DataFrame from the following data:

## Q6. BigMart is one of the biggest retailers in Europe and has operations across multiple countries. You are a data analyst in the IT team of BigMart. Invoice and SKU wise sales data for the Year 2011 is shared with you. You need to prepare meaningful charts to showcase the various sales trends for 2011 to top management. 

## Key issues:
### Data should be displayed pictorially to capture the attention of top management.

## Considerations: NONE

## Data volume:
### Approx. 500K records – file BigMartSalesData.csv

## Additional information: NA

## Business benefits:
### This exercise is an annual exercise and BigMart makes an important investment decision based on trends.

## Approach to solve:
### You must use the fundamentals of Matplotlib covered in module 5 and plot the following fourcharts/graph: 
### 1. Plot total sales per month fortheYear 2011. How have the total sales increased over the months in the Year 2011? Which month has the lowest sales?
### 2. PlotaPie Chart for the Year 2011 Country Wise. Which Country contributes highest towards sales?
### 3. Plot Scatter Plot for the invoice amounts and see the concentration of amount.  In which range have most of the invoice amounts concentrated?

```python
dataset = pd.read_csv('BigMartSalesData.csv', encoding= 'unicode_escape')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
data = dataset.copy()
date = pd.to_datetime(data["InvoiceDate"])
data['Day'] = date.dt.day
data['Month'] = date.dt.month
data['Year'] = date.dt.year
data.head()
```

```python
data = data.drop(data[data.Year!=2011].index)
```

```python
data.head()
```

```python
import numpy as np 

#Plotting the sales data for year 2011

total = data.groupby('Month').agg({'Quantity':np.sum})
print (total)
data.groupby('Month').agg({'Quantity':np.sum}).plot()
plt.xlabel('Year of 2011')
plt.ylabel('Total Sales')
plt.title('Sales per month')
plt.show()

#February has the lowest sales in 2011
```

```python
#Plot Total Sales Per Month for Year 2011 as Bar Chart

data.groupby('Month').agg({'Quantity':np.sum}).plot()
plt.xlabel('Year of 2011')
plt.ylabel('Total Sales')
plt.title('Sales Per Month in 2011')
plt.show()
```

```python
#Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?

plt.figure(figsize=(5,5))
sales_country = data.groupby('Country').agg({'Quantity':np.sum})
sales_country.reset_index(inplace=True)
plt.pie(sales_country.Quantity, labels=sales_country.index,autopct='%1.1f%%')
plt.show()

#United Kingdom contributes highest towards sales with 84.5%. 
```

```python
# Plot Scatter Plot for the invoice amounts and see the concentration of amount. 
# In which range most of the invoice amounts are concentrated

sales_invoice = dataset.groupby('InvoiceNo').agg({'Quantity':np.sum})
print(sales_invoice)
plt.scatter(sales_invoice.values,sales_invoice.values,color='red')
plt.grid(True)
plt.show()

# Most of the datapoints fall under 25000. 
```
