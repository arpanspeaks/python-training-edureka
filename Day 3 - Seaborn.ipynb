```python
from matplotlib import pyplot as plt, style
import seaborn as sns, pandas as pd, numpy as np
```

```python
style.use('ggplot')
```

```python
df = pd.read_csv("Car_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(data=df, x='Horsepower', y='Fuel efficiency')
plt.show()
```

```python
houseData = pd.read_csv("house.csv")
```

```python
houseData.head()
```

```python
mygraph = sns.FacetGrid(houseData,col='Neighborhood',hue='Brick')
mygraph.map(plt.scatter,"SqFt","Price").add_legend()
plt.show()
```

```python
sns.countplot(data=houseData,y="Neighborhood")
plt.show()
```

```python
sns.countplot(data=houseData,y="Brick")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=houseData, x='Neighborhood', y='Price')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(data=houseData, x='Neighborhood', y='Price')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.distplot(houseData.Price)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.histplot(houseData.Price,kde=True)
plt.show()
```
