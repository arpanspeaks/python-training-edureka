## Line Chart

```python
from matplotlib import pyplot as plt, style
import seaborn as sns, pandas as pd, numpy as np
```

```python
plt.plot([12,34,19.6])
plt.show()
```

```python
plt.plot(['Jan', 'Feb', 'Mar'], [12, 34, 19.6])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales by Month")
plt.show()
```

```python
plt.plot(['Jan', 'Feb', 'Mar'], [12, 34, 19.6])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales by Month")
plt.show()
```

```python
df = pd.read_csv("Car_sales.csv")
```

```python
df.head()
```

```python
carSales = df.groupby('Manufacturer')['Sales in thousands'].sum()
```

```python
carSales
```

```python
plt.figure(figsize=(14,6))
plt.plot(carSales)
plt.xlabel("Manufactures")
plt.ylabel("Sales")
plt.title("Sales by Manufactures")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(['Jan', 'Feb', 'Mar'], [12, 34, 19.6], label="2022", color="teal", marker="o", lw=2)
plt.plot(['Jan', 'Feb', 'Mar'], [11, 27, 23], label="2021", linestyle="dashed", marker="o")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales by Month")
plt.legend(loc="best")
plt.show()
```

## Bar Chart

```python
plt.figure(figsize=(14,6))
plt.bar(carSales.index, carSales)
plt.xlabel("Manufactures")
plt.ylabel("Sales")
plt.title("Sales by Manufactures")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.barh(carSales.index, carSales)
plt.xlabel("Manufactures")
plt.ylabel("Sales")
plt.title("Sales by Manufactures")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Vehicle type'].unique()
```

```python
report=pd.pivot_table(data=df,index='Manufacturer',columns='Vehicle type',values='Sales in thousands',aggfunc='sum').fillna(0)
```

```python
plt.figure(figsize=(14,8))
plt.bar(report.index,report.Car)
plt.bar(report.index,report.Passenger,bottom=report.Car)
plt.xlabel("Manufacturer")
plt.ylabel("Sales")
plt.title("Sales by Manufacturer")
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram
plt.figure(figsize=(14,6))
plt.hist(df['Sales in thousands'], bins=20)
plt.show()
```

```python
# Pie Chart
```

```python
vt = df.groupby('Vehicle type')['Sales in thousands'].sum()
```

```python
vt
```

```python
plt.figure(figsize=(14,6))
plt.pie(vt, labels=vt.index, autopct='%.2f')
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.pie(vt, labels=vt.index, autopct='%.2f', explode=[0,0.1])
plt.show()
```

```python
# Scatter plot
```

```python
plt.figure(figsize=(14,6))
plt.scatter(df['Horsepower'], df['Fuel efficiency'])
plt.show()
```

```python
# Saving figure
plt.figure(figsize=(14,8))
plt.barh(carSales.index, carSales)
plt.xlabel("Manufactures")
plt.ylabel("Sales")
plt.title("Sales by Manufactures")
plt.xticks(rotation=45)
plt.savefig("fig.png")
```

```python
# Sub plots
plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
plt.bar(carSales.index, carSales)
plt.xlabel("Manufactures")
plt.ylabel("Sales")
plt.title("Sales by Manufactures")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.pie(vt, labels=vt.index, autopct='%.2f', explode=[0,0.1])

plt.show()
```
